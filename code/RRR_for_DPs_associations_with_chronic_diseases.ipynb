```r
library(readxl)
```

```r
data <- read_xlsx("food_data.xlsx")
data_clusters_gen <- read_xlsx("cluster_data_gen.xlsx")
data_clusters_gen
data_clusters_male <- read_xlsx("cluster_data_male.xlsx")
data_clusters_male
data_clusters_female <- read_xlsx("cluster_data_female.xlsx")
data_clusters_female
```

```r
laboratory_y <- as.matrix(data[,3:9])
food_groups_x <- as.matrix(data[,10:30])
dim(food_groups_x)
dim(laboratory_y)

laboratory_cl_y <- as.matrix(data_clusters_gen[,6:12])
clusters_x <- as.matrix(data_clusters_gen[, c(3:5, 13:14)])
dim(laboratory_cl_y)
dim(clusters_x)

laboratory_cl_y_male <- as.matrix(data_clusters_male[,4:10])
clusters_x_male <- as.matrix(data_clusters_male[, c(3, 11:14)])
dim(laboratory_cl_y_male)
dim(clusters_x_male)

laboratory_cl_y_female <- as.matrix(data_clusters_female[,4:10])
clusters_x_female <- as.matrix(data_clusters_female[, c(3, 11:14)])
dim(laboratory_cl_y_female)
dim(clusters_x_female)
```

```r
(rr_fit <- rrpack::rrr(laboratory_y, food_groups_x, maxrank=10, df.type ='exact', ic.type='AIC'))
(rr_fit_cl_gen <- rrpack::rrr(laboratory_cl_y, clusters_x, maxrank=10, df.type ='exact', ic.type='AIC'))
(rr_fit_cl_male <- rrpack::rrr(laboratory_cl_y_male, clusters_x_male, maxrank=10, df.type ='exact', ic.type='AIC'))
(rr_fit_cl_female <- rrpack::rrr(laboratory_cl_y_female, clusters_x_female, maxrank=10, df.type ='exact', ic.type='AIC'))
```

```r
coef <- rr_fit$coef
colnames(coef) <- colnames(laboratory_y)
levelplot(coef, scales=list(x=list(rot=90)))

coef_cl_gen <- rr_fit_cl_gen$coef
colnames(coef_cl_gen) <- colnames(laboratory_cl_y)
levelplot(coef_cl_gen, scales=list(x=list(rot=90)))

coef_cl_male <- rr_fit_cl_male$coef
colnames(coef_cl_male) <- colnames(laboratory_cl_y_male)
levelplot(coef_cl_male, scales=list(x=list(rot=90)))

coef_cl_female <- rr_fit_cl_female$coef
colnames(coef_cl_female) <- colnames(laboratory_cl_y_female)
levelplot(coef_cl_female, scales=list(x=list(rot=90)))
```

```r
rrr_coefs <- rr_fit$coef.ls
rrr_coefs_df <- as.data.frame(rrr_coefs)
rrr_coefs_df

rrr_coefs_cl <- rr_fit_cl_gen$coef.ls
rrr_coefs_cl_df <- as.data.frame(rrr_coefs_cl)
rrr_coefs_cl_df

rrr_coefs_cl_male <- rr_fit_cl_male$coef.ls
rrr_coefs_cl_df_male <- as.data.frame(rrr_coefs_cl_male)
rrr_coefs_cl_df_male

rrr_coefs_cl_female <- rr_fit_cl_female$coef.ls
rrr_coefs_cl_df_female <- as.data.frame(rrr_coefs_cl_female)
rrr_coefs_cl_df_female
```
